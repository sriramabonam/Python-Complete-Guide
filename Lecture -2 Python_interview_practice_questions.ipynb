```python
# Moderate Level:
# Write a Python program to implement binary search on a sorted list.


def search(sorted_list, n):
    l = 0
    r = len(sorted_list) - 1

    while l <= r:
        mid = (l + r) // 2

        if sorted_list[mid] == n:
            return mid

        if sorted_list[mid] < n:
            l = mid + 1
        else:
            r = mid - 1
    return -1

lst = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
n = 8
target_element = 16

loc = search(lst, n)

if loc != -1:
    print(f"{n} found at index {loc}")
else:
    print(f"{n} not found in the list")




# Write a Python program to remove duplicates from a list.

def rem_dup(ls):
    dup=[]
    uni=[]
    for i in ls:
        if i in uni:
            dup.append(i)
        else:
            uni.append(i)
    return uni
ls=[1,2,3,2,4,6,6,7,1]
res=rem_dup(ls)
print(res)

unique=list(set(lst))



# Write a Python function to calculate the Fibonacci sequence up to a given number.

def fibonacci(n):
    if n ==0 or n==1:
        return n
    else:
        return fibonacci(n-1)+fibonacci(n-2)


# Write a Python program to merge two sorted lists into a single sorted list.

def mergelst(l1,l2):
    if l1[-1]<l2[0]:
        l1.extend(l2)
        return l1
    elif l2[-1] <l1[0]:
        l2.extend(l1)
        return l2

def merg_sorted_list(list1,list2)
	merged_list = sorted(list1 + list2)
	return merged_list

list1 = [1,3,5,7]
list2 = [2,4,6,8]

print(merg_sorted_list(list1,list2))


```
